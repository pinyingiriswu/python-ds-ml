```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
#Multiplots
plt.subplot(1,2,1) #(row, column, which column) #edit 1st column
plt.plot(x,y,'r') #red line

plt.subplot(1,2,2)#edit 2nd column
plt.plot(y,x,'b') #blue line
```

```python
#Object-Oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #add axes at percentage of fraction (ranges from 0 to 1) #left,bottom,width,height
axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #inside

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
#Create subplots
fig,axes = plt.subplots(nrows=1,ncols=2) #axes is a list of matplotlib axes

#You can iterate through axes since it is a list
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout() #fix overlaps
```

```python
axes
```

```python
#Figure Size and DPI(dots per inch)
```

```python
fig = plt.figure(figsize=(8,2),dpi=100) #figsize=(width,height)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2)) #figsize=(width,height)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
#Legends
```

```python
fig = plt.figure() 

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

#ax.legend(loc=0) #loc=0 means best location
#ax.legend(loc=10) #loc=10 means center
ax.legend(loc=(0.1,0.1)) #specify legend location with tuple
```

```python
#Plot Appearance
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,y,color='#FF8C00') #RGB Hex Code
#ax.plot(x,y,color='purple',linewidth=3,alpha=0.5) #alpha is for transparency
#ax.plot(x,y,color='purple',lw=3,linestyle='-.')
ax.plot(x,y,color='purple',lw=3,ls='--',marker='s',markersize=20,
     markerfacecolor='yellow',markeredgewidth=3, markeredgecolor='green') #marker marks the points
```

```python
#Set x limit and y limit
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')

ax.set_xlim([0,1]) #show the plot between x=0 and x=1
ax.set_ylim([0,2]) #show the plot between y=0 and y=2
```
