```python
#Categorical Plots
```

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) #default estimator is mean
```

```python
sns.countplot(x='sex',data=tips)
```

```python
#sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
sns.boxplot(x='day',y='total_bill',data=tips) #suited for management team to present
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True) #jitter set to True show the density better
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips) #combination of stripplot and violinplot
#suggest not to use large datasets for swarmplot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black') #more suited for data scientists
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar') #general method to set a plot
```

```python
#Matrix Plots
```

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
tc
```

```python
#Matrix Plot type 1 - Heatmap
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm') #annot declares numerical values
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=3)
#Heatmap displays data in the order you defined
```

```python
#Matrix Plot type 2 - Cluster Map
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1) #standard_scale enables you to set your custom scale
#Cluster map groups data out of similarity
```

```python
#Grid
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
#Pair Grid
g = sns.PairGrid(iris) #creates empty pair grid so you can customize your own plots
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
#Facet Grid
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker') #Creates empty grid
g.map(sns.distplot,'total_bill') #need 1 argument
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip') #need 2 arguments
```
